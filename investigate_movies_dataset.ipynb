## Step 3: Filling the zero values:
## We can achieve it by taking the mean of the column and replacing the zero values with the mean calculated:

```python
#Check if the revenue_adj column contain any 0 values
df['revenue_adj'].value_counts()
```

```python
#Filling the zero values with the mean for 'revenue_adj'
rev_adj_mean=df['revenue_adj'].mean()
df['revenue_adj'].replace(0,rev_adj_mean,inplace=True)
```

```python
# Zero Values Fixed
df['revenue_adj'].value_counts()
```

```python
df['budget_adj'].value_counts()
```

```python
df['budget_adj'].replace(0,df['budget_adj'].mean(),inplace=True)
```

```python
df['budget_adj'].value_counts()
```

## Step4 : Removing The Extraneous Columns From The Dataset:

### Since, 'keywords' and 'overview' columns provide no meaningful information to carry out analysis, it's a better idea to drop the columns. Also, they take up unnecessary space in the dataset, adding up to the overall data.

```python
df.drop(['keywords','overview','tagline','production_companies','cast','homepage'],axis=1,inplace=True)
```

## Drop all the null values in column genre if any present:

```python
df.dropna(subset=['genres'],axis=0,inplace=True)
```

## Split the column 'genres' on the basis of a seperator - '|' and label each column :

```python
new_genre=df['genres'].str.split('|',expand=True)
labels={0:'genre1',1:'genre2',2:'genre3',3:'genre4',4:'genre5'}
new_genre.rename(labels,axis=1,inplace=True)
new_genre
```

```python
# Merging the newly formed columns with the origonal dataframe:
frames=[new_genre,df]
df=pd.concat(frames,join='inner',ignore_index=False,axis=1)
df
```

```python
# Dropping the original 'Genres' column :
df.drop('genres',axis=1,inplace=True)
```

```python
df.info()
```

# Data Exploration Using Visuals :

## 4] Which are the Top 5 Revenue Generating Genres:  

```python
# Finding the total count of each genre and then fetching top 5 genres among them:
rev_total=df.groupby(['genre1'])['revenue_adj'].sum()+df.groupby(['genre2'])['revenue_adj'].sum()+df.groupby(['genre3'])['revenue_adj'].sum()+df.groupby(['genre4'])['revenue_adj'].sum()+df.groupby(['genre5'])['revenue_adj'].sum()
rev_total.nlargest(5)
```

## 5] Represent Revenue-wise Distribution of Top 5 Movie Genres Using a Bar Chart :

```python
# Calculate genre-wise total revenue collected and then extract the top 5 revenue generating movies
adventure=df[(df['genre1']=='Adventure') | (df['genre2']=='Adventure') | (df['genre3']=='Adventure') | (df['genre4']=='Adventure') | (df['genre5']=='Adventure')]
adventure_sum=adventure['revenue_adj'].sum()
adventure_sum 
```

```python
thriller=df[(df['genre1']=='Thriller') | (df['genre2']=='Thriller') | (df['genre3']=='Thriller') | (df['genre4']=='Thriller') | (df['genre5']=='Thriller')]
thriller_sum=thriller['revenue_adj'].sum()
thriller_sum
```

```python
action=df[(df['genre1']=='Action') | (df['genre2']=='Action') | (df['genre3']=='Action') | (df['genre4']=='Action') | (df['genre5']=='Action')]
action_sum=action['revenue_adj'].sum()
action_sum
```

```python
drama=df[(df['genre1']=='Drama') |(df['genre2']=='Drama') | (df['genre3']=='Drama') | (df['genre4']=='Drama') | (df['genre5']=='Drama')]
drama_sum=drama['revenue_adj'].sum()
drama_sum
```

```python
comedy=df[(df['genre1']=='Comedy') | (df['genre2']=='Comedy') | (df['genre3']=='Comedy') | (df['genre4']=='Comedy') | (df['genre5']=='Comedy')]
comedy_sum=comedy['revenue_adj'].sum()
comedy_sum
```

```python
x_axs=['Adventure','Comedy','Drama','Thriller','Action']
y_axs=[adventure_sum,comedy_sum,drama_sum,thriller_sum,action_sum]
plt.title('Movie Genre vs Total Revenue')
plt.xlabel('Movie Genres')
plt.ylabel('Total Revenue')
plt.bar(x_axs,y_axs)
```

## 6] Represent percentage-wise Distribution of Top 5 Movie Genres Using a Pie-Chart :


```python
df_pie=pd.DataFrame({'genres':['adventure','comedy','drama','thriller','action'],
                     'revenue':[adventure_sum,comedy_sum,drama_sum,thriller_sum,action_sum]},
                    index=['ADVENTURE','COMEDY','DRAMA','THRILLER','ACTION'])
df_pie.plot.pie(y='revenue',figsize=(12,12),autopct='%1.0f%%')
```

## 7] Discover the co-relation between 'revenue' and 'vote_count'

```python
x= df['vote_count']
y = df['revenue']
plt.figure(figsize=(10, 10), dpi=80)
plt.scatter(x,y)
plt.title("Vote Count vs Revenue")
plt.xlabel("Vote Count")
plt.ylabel("Revenue (in million of dollars)")
plt.show()
```
